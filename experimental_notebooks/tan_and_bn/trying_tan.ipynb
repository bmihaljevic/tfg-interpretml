```python
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
import pyAgrum.skbn as skbn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = pd.read_csv("data/X_train.csv")
X_test = pd.read_csv("data/X_test.csv")
y_train = pd.read_csv("data/y_train.csv")
y_test = pd.read_csv("data/y_test.csv")
```

```python
TAN = skbn.BNClassifier(learningMethod="Chow-Liu", prior= 'Smoothing', priorWeight = 0.5,
                          discretizationStrategy = 'quantile', usePR = True, significant_digit = 13)
```

```python
TAN.fit(X_train, y_train)
```

```python
y_test.squeeze()
```

```python
(TAN.predict(X_test) == y_test.squeeze()).sum() / len(y_test)
```

```python
gnb.sideBySide(TAN.bn,gnb.getInference(TAN.bn,size='15!'))
```
